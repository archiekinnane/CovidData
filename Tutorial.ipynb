# Graphing county and state coronavirus data

First, run the following code so your plots show up in your Jupyter notebook.

```python
%matplotlib inline
```

Once you've downloaded the package from GitHub and saved it on your machine in a path that Python will find, then you can use it to graph Covid-19 data for any state or county.

If you've downloaded the package folder but Python is saying it can't find it, search your disk for 'numpy,' and put the 'covid_g' folder in the same directory.

```python
from covid_g import state
```

Graph your state. You can initialize the function with your desired state, or, if left blank, it will ask you for input.

```python
state.graph('NY')
```

If you leave the parameter blank, it will prompt you for input.

```python
state.graph()
```

To graph a county, import the 'county' module from the 'covid_g' package.

```python
from covid_g import county
```

Again, you can initialize the function with county.graph('county', 'state'), or leave blank and be prompted for input

```python
county.graph('New York City', 'New York')
```

```python
county.graph()
```
